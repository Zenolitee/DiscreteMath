```python
import numpy as np

def parse_arff(file_path):
    attributes, data, nominal_mapping = [], [], []

    with open(file_path, 'r') as file:
        data_started = False
        for line in file:
            line = line.strip()

            if not line or line.startswith(('%', '@relation')):
                continue

            if line.startswith('@attribute'):
                attr_name, attr_type, attr_values = parse_attribute_line(line)
                attributes.append((attr_name, attr_type, attr_values))
                if attr_type == 'nominal':
                    nominal_mapping.append({value: index for index, value in enumerate(attr_values)})

            if line.startswith('@data'):
                data_started = True
                continue

            if data_started:
                data.append(line.split(','))

    return attributes, data, nominal_mapping

def parse_attribute_line(line):
    parts = line.split()
    attr_name = parts[1].strip()
    if '{' in line:  # Nominal data
        values = line[line.index('{') + 1:line.index('}')].split(',')
        return attr_name, 'nominal', values
    else:  # Numeric data
        return attr_name, 'numeric', None

def process_data(attributes, data, nominal_mapping):
    for row in data:
        for i, (attr_name, attr_type, _) in enumerate(attributes):
            if attr_type == 'nominal' and i < len(nominal_mapping):
                try:
                    row[i] = nominal_mapping[i].get(row[i], row[i])
                except TypeError:
                    row[i] = 1
            elif attr_type == 'numeric':
                try:
                    row[i] = float(row[i])
                except Exception:
                    row[i] = 1

    return data

def display_matrix(matrix):
    for row in matrix:
        print("|".join(f"{str(cell):^10}" for cell in row))

# Example usage
file_path = r'C:\Users\Zen\Desktop\Panganiban-Prefi\V4 data\2017 Q1.arff' #change this to whatever in the V4 folder.
attributes, data, nominal_mapping = parse_arff(file_path)
processed_data = process_data(attributes, data, nominal_mapping)
display_matrix(processed_data)
```
